```python
import random
print("Welcome to Number Guessing Game!")
print("Guess number between 1 to 100 ")

secret_number=random.randint(1,100)

for i in range(0,7):
    try :
     user_input=int(input("Enter a number: "))
     if user_input<1 or user_input>100:
        print("Invalid")
     elif secret_number < user_input:
        print("Too high")
     elif secret_number > user_input:
        print("Too low")
     else : 
        print("Your guess is correct")
        break

    except ValueError:
        print("Invalid! Guess number between 1 to 100")

if secret_number == user_input:
    print(f"Your have guessed the secret number {secret_number} in {i+1} attempts!")
else:
    print("Sorry! Your attempts are Completed. Try Again.")
```
